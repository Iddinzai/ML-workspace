```python
cars = pd.read_csv("car-sales.csv")
cars["Price"] = cars["Price"].str.replace('[\\$,]', '',regex=True).astype(float).astype(int)
cars
```

```python
cars["sale date"] = pd.date_range("1/1/2020", periods=len(cars))
cars["total sales"] = cars["Price"].astype(int).cumsum()
cars
```

```python
cars.plot(x="sale date" , y="total sales");
```
